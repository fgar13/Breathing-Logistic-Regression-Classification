```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
import glob
from os import listdir
from os.path import isfile, join
from csv import writer
```

```python
with open(files_with_full_path_name[2]) as f:
    print(f)
```

```python
pandas_data_frames_to_concatenate = []
for file in files_with_full_path_name: 
    pandas_data_frame = pd.read_excel(file)
    pandas_data_frames_to_concatenate.append(pandas_data_frame)
concatenated_pandas_data_frames = pd.concat(pandas_data_frames_to_concatenate)
concatenated_pandas_data_frames
```

```python
new_table = concatenated_pandas_data_frames
axis_numbers = [number for number in range(0,len(concatenated_pandas_data_frames))] # len(concatenated_pandas_data_frames.columns) for columns, default is rows
new_table = new_table.set_axis(axis_numbers) #fix duplicate numbers along row axis, default for set_axis is rows
axis_numbers = [number for number in range(0,109)] #all features, "slice" [:78] is all the positions, and [79:] is filters
new_table = new_table.set_axis(axis_numbers, axis='columns') #when changing default, axis="columns"
new_table = new_table.loc[165:176,0:76] #columns includes frame, features, and class
new_table
```

```python
X = [] #input
for  index, row in new_table.iterrows():
    X.append(row)
print(X[3])
Y = [] #classes
```

```python
X = [] #input of features
Y = [] #classes
for row in new_table.itertuples():
    X.append(list(row[1:77])) #excluding 78, because 78 is the class label, and excluding 0 because it is redudant with 1
    #also, list() turns tuple into list
    Y.append(row[-1:][0]) #class label only
print(X)
print("\n\n\n\n")
print(Y)
```

```python
clf = LogisticRegression(random_state=0).fit(X, Y)
clf.predict(X[:])
clf.score(X, Y)
```
